```sql
SELECT  i_item_id
        ,i_item_desc
        ,i_current_price
FROM samples.tpcds_sf1000.item
,       samples.tpcds_sf1000.inventory
,       samples.tpcds_sf1000.date_dim
,       samples.tpcds_sf1000.catalog_sales
WHERE   i_current_price BETWEEN 33 AND 33 + 30
AND     inv_item_sk = i_item_sk
AND     d_date_sk = inv_date_sk
AND     d_date BETWEEN CAST('1999-06-20' AS DATE) AND (
            CAST('1999-06-20' AS DATE) + interval'60'day
)
AND     i_manufact_id IN (882,737,821,742)
AND     inv_quantity_on_hand BETWEEN 100 AND 500
AND     cs_item_sk = i_item_sk
GROUP BY i_item_id
          ,i_item_desc
          ,i_current_price
ORDER BY i_item_id
LIMIT   100
;
```
