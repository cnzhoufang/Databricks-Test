```sql
SET use_cached_result = false;
SELECT  i_item_id
        ,i_item_desc
        ,i_category
        ,i_class
        ,i_current_price
        ,SUM(cs_ext_sales_price) AS itemrevenue
        ,SUM(cs_ext_sales_price) * 100 / SUM(SUM(cs_ext_sales_price)) OVER (PARTITION BY i_class ) AS revenueratio
FROM workspace.tpcds_sf1000_iceberg.catalog_sales
,       workspace.tpcds_sf1000_iceberg.item
,       workspace.tpcds_sf1000_iceberg.date_dim
WHERE   cs_item_sk = i_item_sk
AND     i_category IN ('Home','Electronics','Men')
AND     cs_sold_date_sk = d_date_sk
AND     d_date BETWEEN CAST('1999-04-02' AS DATE) AND (
            CAST('1999-04-02' AS DATE) + interval'30'day
)
GROUP BY i_item_id
          ,i_item_desc
          ,i_category
          ,i_class
          ,i_current_price
ORDER BY i_category,i_class,i_item_id,i_item_desc,revenueratio
LIMIT   100
;
```
