```sql
SET use_cached_result = false;
SELECT  a.ca_state state
        ,COUNT(*) cnt
FROM workspace.tpcds_sf1000_iceberg.customer_address a
,       workspace.tpcds_sf1000_iceberg.customer c
,       workspace.tpcds_sf1000_iceberg.store_sales s
,       workspace.tpcds_sf1000_iceberg.date_dim d
,       workspace.tpcds_sf1000_iceberg.item i
WHERE   a.ca_address_sk = c.c_current_addr_sk
AND     c.c_customer_sk = s.ss_customer_sk
AND     s.ss_sold_date_sk = d.d_date_sk
AND     s.ss_item_sk = i.i_item_sk
AND     d.d_month_seq = (
            SELECT  DISTINCT (
                        d_month_seq
            )
            FROM workspace.tpcds_sf1000_iceberg.date_dim
            WHERE   d_year = 2001
            AND     d_moy = 1
        ) 
AND     i.i_current_price > 1.2 * (
            SELECT  AVG(j.i_current_price)
            FROM workspace.tpcds_sf1000_iceberg.item j
            WHERE   j.i_category = i.i_category
        ) 
GROUP BY a.ca_state
HAVING  COUNT(*) >= 10
ORDER BY cnt,a.ca_state
LIMIT   100
;
```
