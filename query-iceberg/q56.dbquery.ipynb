```sql
SET use_cached_result = false;
WITH ss AS 
(
    SELECT  i_item_id
            ,SUM(ss_ext_sales_price) total_sales
    FROM samples.tpcds_sf1000.store_sales
    ,       samples.tpcds_sf1000.date_dim
    ,       samples.tpcds_sf1000.customer_address
    ,       samples.tpcds_sf1000.item
    WHERE   i_item_id IN (
                SELECT  i_item_id
                FROM samples.tpcds_sf1000.item
                WHERE   i_color IN ('floral','burnished','bisque')
            ) 
    AND     ss_item_sk = i_item_sk
    AND     ss_sold_date_sk = d_date_sk
    AND     d_year = 2000
    AND     d_moy = 5
    AND     ss_addr_sk = ca_address_sk
    AND     ca_gmt_offset = -6
    GROUP BY i_item_id
)
,cs AS 
(
    SELECT  i_item_id
            ,SUM(cs_ext_sales_price) total_sales
    FROM samples.tpcds_sf1000.catalog_sales
    ,       samples.tpcds_sf1000.date_dim
    ,       samples.tpcds_sf1000.customer_address
    ,       samples.tpcds_sf1000.item
    WHERE   i_item_id IN (
                SELECT  i_item_id
                FROM samples.tpcds_sf1000.item
                WHERE   i_color IN ('floral','burnished','bisque')
            ) 
    AND     cs_item_sk = i_item_sk
    AND     cs_sold_date_sk = d_date_sk
    AND     d_year = 2000
    AND     d_moy = 5
    AND     cs_bill_addr_sk = ca_address_sk
    AND     ca_gmt_offset = -6
    GROUP BY i_item_id
)
,ws AS 
(
    SELECT  i_item_id
            ,SUM(ws_ext_sales_price) total_sales
    FROM samples.tpcds_sf1000.web_sales
    ,       samples.tpcds_sf1000.date_dim
    ,       samples.tpcds_sf1000.customer_address
    ,       samples.tpcds_sf1000.item
    WHERE   i_item_id IN (
                SELECT  i_item_id
                FROM samples.tpcds_sf1000.item
                WHERE   i_color IN ('floral','burnished','bisque')
            ) 
    AND     ws_item_sk = i_item_sk
    AND     ws_sold_date_sk = d_date_sk
    AND     d_year = 2000
    AND     d_moy = 5
    AND     ws_bill_addr_sk = ca_address_sk
    AND     ca_gmt_offset = -6
    GROUP BY i_item_id
)
SELECT  i_item_id
        ,SUM(total_sales) total_sales
FROM    (
            SELECT  *
            FROM ss
            UNION ALL
            SELECT  *
            FROM cs
            UNION ALL
            SELECT  *
            FROM ws
        ) tmp1
GROUP BY i_item_id
ORDER BY total_sales,i_item_id
LIMIT   100
;
```
