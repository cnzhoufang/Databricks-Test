```sql
SET use_cached_result = false;
SELECT  i_item_id
        ,i_item_desc
        ,i_current_price
FROM samples.tpcds_sf1000.item
,       samples.tpcds_sf1000.inventory
,       samples.tpcds_sf1000.date_dim
,       samples.tpcds_sf1000.store_sales
WHERE   i_current_price BETWEEN 22 AND 22 + 30
AND     inv_item_sk = i_item_sk
AND     d_date_sk = inv_date_sk
AND     d_date BETWEEN CAST('2001-02-22' AS DATE) AND (
            CAST('2001-02-22' AS DATE) + interval'60'day
)
AND     i_manufact_id IN (624,453,961,174)
AND     inv_quantity_on_hand BETWEEN 100 AND 500
AND     ss_item_sk = i_item_sk
GROUP BY i_item_id
          ,i_item_desc
          ,i_current_price
ORDER BY i_item_id
LIMIT   100
;
```
