```sql
SET use_cached_result = false;
SELECT  i_item_id
        ,i_item_desc
        ,i_category
        ,i_class
        ,i_current_price
        ,SUM(ws_ext_sales_price) AS itemrevenue
        ,SUM(ws_ext_sales_price) * 100 / SUM(SUM(ws_ext_sales_price)) OVER (PARTITION BY i_class ) AS revenueratio
FROM samples.tpcds_sf1000.web_sales
,       samples.tpcds_sf1000.item
,       samples.tpcds_sf1000.date_dim
WHERE   ws_item_sk = i_item_sk
AND     i_category IN ('Jewelry','Electronics','Children')
AND     ws_sold_date_sk = d_date_sk
AND     d_date BETWEEN CAST('2000-03-25' AS DATE) AND (
            CAST('2000-03-25' AS DATE) + interval'30'day
)
GROUP BY i_item_id
          ,i_item_desc
          ,i_category
          ,i_class
          ,i_current_price
ORDER BY i_category,i_class,i_item_id,i_item_desc,revenueratio
LIMIT   100
;
```
