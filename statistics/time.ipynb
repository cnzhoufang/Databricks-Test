```sql
SELECT
  workspace_id,
  statement_text,
  from_utc_timestamp(start_time, 'Asia/Shanghai') AS start_time_shanghai,
  from_utc_timestamp(end_time, 'Asia/Shanghai') AS end_time_shanghai,
  total_duration_ms / 1000 AS total_duration_sec,
  waiting_for_compute_duration_ms / 1000 AS waiting_for_compute_duration_sec,
  waiting_at_capacity_duration_ms / 1000 AS waiting_at_capacity_duration_sec,
  execution_duration_ms / 1000 AS execution_duration_sec,
  compilation_duration_ms / 1000 AS compilation_duration_sec,
  total_task_duration_ms / 1000 AS total_task_duration_sec,
  result_fetch_duration_ms / 1000 AS result_fetch_duration_sec
FROM
  system.query.history
WHERE
  -- workspace_id = '7474650251531711'
  start_time >= to_utc_timestamp(:start1, 'Asia/Shanghai')
  AND start_time <= to_utc_timestamp(:start2, 'Asia/Shanghai')
  AND statement_text != 'self._target(*self._args, **self._kwargs)'
  AND statement_text != 'SET use_cached_result = false'
ORDER BY
  start_time
```
